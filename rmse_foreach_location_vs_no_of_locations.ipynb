```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
import os
```

```python
base_dir ='data\AirQuality\AirQo'
number_of_features_vs_rmse_for_different_locations= pd.read_csv(os.path.join(base_dir,'number_of_features_vs_rmse_for_different_locations.csv'))
number_of_features_vs_rmse_for_different_locations_a = pd.read_csv(os.path.join(base_dir,'number_of_features_vs_rmse_for_different_locations_a.csv'))
number_of_features_vs_rmse_for_different_locations.head()
```

```python
    plt.rcParams.update({'font.size':14, 'font.weight':'bold'})
    x_axis_label = 'Number of Features'
    y_axis_label =  'RMSE'
    chart_title  =  'RMSE vs. Number of features for selected locations '
    file_path    =   os.path.join(base_dir , 'RMSE_vs_Number_of_Features.png')
    
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111)
    plt.plot(number_of_features_vs_rmse_for_different_locations['Number of Features'].values, 
             number_of_features_vs_rmse_for_different_locations['RMSE_Lubowa'].values, 
             color='deeppink', marker='d', linestyle='solid',  label='Lubowa', linewidth=1.5,markersize=12)
    
    plt.plot(number_of_features_vs_rmse_for_different_locations['Number of Features'].values, 
             number_of_features_vs_rmse_for_different_locations['RMSE_Rubaga'].values, 
             #Rubaga
             color='black', marker='P', linestyle='solid',  label='Rubaga', linewidth=2.0,markersize=12)
    
    plt.plot(number_of_features_vs_rmse_for_different_locations['Number of Features'].values, 
             number_of_features_vs_rmse_for_different_locations['RMSE_Kireka'].values, #Kireka
             color='blue', marker='o', linestyle='solid',  label='Kireka', linewidth=2.0,markersize=12)
    
    plt.plot(number_of_features_vs_rmse_for_different_locations['Number of Features'].values, 
             number_of_features_vs_rmse_for_different_locations['RMSE_JinjaRoad'].values, #Jinja Road
             color='darkcyan', marker='*', linestyle='solid',  label='Jinja Road',linewidth=2.0, markersize=12)
    
    plt.plot(number_of_features_vs_rmse_for_different_locations['Number of Features'].values, 
             number_of_features_vs_rmse_for_different_locations['RMSE_Nsambya'].values, #Nsambya
             color='midnightblue', marker='p', linestyle='dashed',  label='Nsambya ',linewidth=2.0, markersize=12)
    
    #plt.title(chart_title, fontsize=20)
    plt.ylabel(y_axis_label, fontsize=20)
    plt.xlabel(x_axis_label,fontsize=20)
    
    #plt.xticks(rotation=60)
    plt.legend()
    plt.tight_layout()
    plt.show()
    fig.savefig(file_path)
```
